```python
import pandas as pd
import altair as alt
```

```python
import os
import requests
url='https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/licenses_fall2022.csv'

response = requests.get(url)
with open('data.csv', 'wb') as f:
    f.write(response.content)
df = pd.read_csv('data.csv')
df
```

```python
print("Dataset Overview:")
print(df.info())
```

```python
print("\nVisualization 1: Top 10 License Types")
print("This bar chart shows the top 10 license types by their count.")


top_license_types = (df["License Type"].value_counts().head(10).reset_index().rename(columns={"index": "License Type", "Description": "License Type"}))

print("\nTop 10 License Types Data:")
print(top_license_types)

bar_chart = (
    alt.Chart(top_license_types)
    .mark_bar()
    .encode(
        y=alt.Y("count:Q", title="Number of Licenses"),
        x=alt.X("License Type:N", sort="-x", title="License Type"),
        color=alt.Color("count:Q", legend=None),
        tooltip=["License Type:N", "count:Q"],
    )
    .properties(
        width=600,
        height=400,
        title="Top 10 License Types"
    )
)

bar_chart.display()
```

```python
df["Effective Date"] = pd.to_datetime(df["Effective Date"])

top_cities = (
    df["City"]
    .value_counts()
    .head(10)
    .index
)

df_top_cities = df[df["City"].isin(top_cities)]


licenses_by_date_city = (
    df_top_cities.groupby(["Effective Date", "City"])
    .size()
    .reset_index(name="count")
)

shape_select = alt.binding_select(
    options=[
        "arrow",
        "circle",
        "square",
        "cross",
        "diamond",
        "triangle",
        "triangle-up",
        "triangle-down",
        "triangle-right",
        "triangle-left",
        "wedge",
        "stroke",
        "M-1,-1H1V1H-1Z",
        "M0,.5L.6,.8L.5,.1L1,-.3L.3,-.4L0,-1L-.3,-.4L-1,-.3L-.5,.1L-.6,.8L0,.5Z",
    ],
    name="shape",
)
shape_var = alt.param(bind=shape_select, value="circle")

angle_slider = alt.binding_range(min=-360, max=360, step=1, name="angle")
angle_var = alt.param(bind=angle_slider, value=0)

size_slider = alt.binding_range(min=0, max=500, step=10, name="size")
size_var = alt.param(bind=size_slider, value=50)

strokeWidth_slider = alt.binding_range(min=0, max=10, step=0.5, name="strokeWidth")
strokeWidth_var = alt.param(bind=strokeWidth_slider, value=2)

brush = alt.selection_point(name="point", encodings=["color"], bind="legend")

scatter_plot = (
    alt.Chart(licenses_by_date_city)
    .mark_point(shape=shape_var,
    angle=angle_var,
    size=size_var,
    strokeWidth=strokeWidth_var,)
    .encode(
        x=alt.X("Effective Date:T", title="Effective Date"),  
        y=alt.Y("count:Q", title="Number of Licenses"),  
        color=alt.condition(brush, "City:N", alt.value('rgba(255, 0, 0, 0)'),title="City"), 
        tooltip=["Effective Date:T", "City:N", "count:Q"],  
    )
    .properties(
        width=800,
        height=400,
        title="Top 10 Cities Issued Licenses Over Time"
    ).add_params(
    shape_var, angle_var, size_var, strokeWidth_var, brush
)
)
scatter_plot
```

```python
bar_chart.save('bar_chart.html')
scatter_plot.save('scatter_plot.html')
```
